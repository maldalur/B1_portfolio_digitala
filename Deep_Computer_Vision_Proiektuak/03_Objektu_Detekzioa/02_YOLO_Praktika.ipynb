## 5ï¸âƒ£ Irudi Simulatuetan Detekzioa

````python
# Detekzio simulazioa
def simulatu_yolo_detekzioa():
    """
    YOLO detekzio simulazioa irudi batean
    """
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    
    # Irudi simulatua (kale baten eszena)
    img = np.random.rand(600, 800, 3) * 0.4 + 0.3
    ax.imshow(img)
    
    # Detekzio simulatuak
    detekzioak = [
        ((100, 150, 120, 200), 'persona', 0.95, 'red'),
        ((250, 180, 100, 150), 'persona', 0.89, 'red'),
        ((450, 200, 200, 150), 'auto', 0.92, 'blue'),
        ((600, 250, 80, 100), 'txakurra', 0.87, 'green'),
        ((50, 50, 100, 80), 'bizikleta', 0.83, 'orange')
    ]
    
    for (x, y, w, h), label, conf, color in detekzioak:
        # Bounding box
        rect = patches.Rectangle(
            (x, y), w, h,
            linewidth=3, edgecolor=color, facecolor='none'
        )
        ax.add_patch(rect)
        
        # Label eta confidence
        ax.text(x, y-10, f'{label} {conf:.2f}',
               bbox=dict(facecolor=color, alpha=0.8),
               fontsize=11, color='white', fontweight='bold')
    
    ax.set_title('ðŸŽ¯ YOLO Detekzio Simulazioa', 
                fontsize=18, fontweight='bold', pad=20)
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    
    print(f"âœ… {len(detekzioak)} objektu detektatuta!")

simulatu_yolo_detekzioa()
````

## 7ï¸âƒ£ Confidence Threshold eta NMS Parametroak

````python
# Threshold-en eragina
def bistaratu_threshold_eragina():
    """
    Threshold desberdinen eragina bistaratu
    """
    # Detekzio simulatuak confidence desberdinarekin
    detekzioak_guztiak = [
        ('persona', 0.95),
        ('auto', 0.92),
        ('persona', 0.89),
        ('txakurra', 0.87),
        ('bizikleta', 0.83),
        ('katu', 0.78),
        ('moto', 0.65),
        ('bankua', 0.52),
        ('landare', 0.45),
        ('taza', 0.38)
    ]
    
    thresholds = [0.5, 0.7, 0.85]
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('ðŸŽ¯ Confidence Threshold-en Eragina', 
                 fontsize=16, fontweight='bold')
    
    for idx, threshold in enumerate(thresholds):
        ax = axes[idx]
        
        # Threshold-a gainditzen duten detekzioak
        detekzio_onartuak = [(obj, conf) for obj, conf in detekzioak_guztiak 
                             if conf >= threshold]
        
        if detekzio_onartuak:
            objektuak = [obj for obj, _ in detekzio_onartuak]
            confidences = [conf for _, conf in detekzio_onartuak]
            colors = plt.cm.RdYlGn([c for c in confidences])
            
            bars = ax.barh(objektuak, confidences, color=colors, edgecolor='black', linewidth=1.5)
        
        ax.axvline(x=threshold, color='red', linestyle='--', linewidth=2, 
                  label=f'Threshold={threshold}')
        ax.set_xlabel('Confidence', fontsize=12)
        ax.set_title(f'Threshold={threshold}\n({len(detekzio_onartuak)} detekzio)', 
                    fontsize=12, fontweight='bold')
        ax.set_xlim([0, 1])
        ax.legend()
        ax.grid(True, alpha=0.3, axis='x')
    
    plt.tight_layout()
    plt.show()

bistaratu_threshold_eragina()
print("âœ… Threshold-en eragina bistaratuta!")
````

## 9ï¸âƒ£ Performance Metrikak

````python
# YOLO performance simulazioa
def simulatu_yolo_performance():
    """
    YOLO eredu desberdinen performance konparaketa
    """
    ereduak = ['YOLOv8n', 'YOLOv8s', 'YOLOv8m', 'YOLOv8l', 'YOLOv8x']
    mAP = [0.37, 0.44, 0.50, 0.53, 0.54]  # mAP@0.5:0.95
    fps = [150, 100, 70, 45, 30]  # Frames per second
    parametroak = [3.2, 11.2, 25.9, 43.7, 68.2]  # Milioi parametro
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('ðŸ“Š YOLOv8 Ereduen Konparaketa', fontsize=16, fontweight='bold')
    
    # mAP
    axes[0].bar(ereduak, mAP, color='skyblue', edgecolor='black', linewidth=1.5)
    axes[0].set_ylabel('mAP@0.5:0.95', fontsize=12)
    axes[0].set_title('ðŸŽ¯ Zehaztasuna', fontsize=14, fontweight='bold')
    axes[0].grid(True, alpha=0.3, axis='y')
    axes[0].set_ylim([0, 0.6])
    
    # FPS
    axes[1].bar(ereduak, fps, color='lightgreen', edgecolor='black', linewidth=1.5)
    axes[1].set_ylabel('FPS', fontsize=12)
    axes[1].set_title('âš¡ Abiadura', fontsize=14, fontweight='bold')
    axes[1].grid(True, alpha=0.3, axis='y')
    
    # Parametroak
    axes[2].bar(ereduak, parametroak, color='lightcoral', edgecolor='black', linewidth=1.5)
    axes[2].set_ylabel('Parametroak (M)', fontsize=12)
    axes[2].set_title('ðŸ’¾ Ereduaren Tamaina', fontsize=14, fontweight='bold')
    axes[2].grid(True, alpha=0.3, axis='y')
    
    for ax in axes:
        ax.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

simulatu_yolo_performance()
print("âœ… Performance konparaketa bistaratuta!")
````

````python
# Entrenamenduko progresio simulazioa
def simulatu_entrenamendua():
    """
    Entrenamenduko metriken progresio simulazioa
    """
    epochs = np.arange(1, 101)
    train_loss = 5.0 * np.exp(-epochs/30) + 0.5 + np.random.rand(100) * 0.1
    val_loss = 5.2 * np.exp(-epochs/30) + 0.7 + np.random.rand(100) * 0.15
    mAP = 0.8 * (1 - np.exp(-epochs/20)) + np.random.rand(100) * 0.05
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle('ðŸ“ˆ YOLO Entrenamenduko Progresio Simulazioa', 
                 fontsize=16, fontweight='bold')
    
    # Loss
    axes[0].plot(epochs, train_loss, label='Train Loss', linewidth=2, color='blue')
    axes[0].plot(epochs, val_loss, label='Val Loss', linewidth=2, color='red')
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Loss', fontsize=12)
    axes[0].set_title('ðŸ“‰ Loss', fontsize=14, fontweight='bold')
    axes[0].legend(fontsize=11)
    axes[0].grid(True, alpha=0.3)
    
    # mAP
    axes[1].plot(epochs, mAP, linewidth=2, color='green')
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('mAP@0.5:0.95', fontsize=12)
    axes[1].set_title('ðŸŽ¯ mAP', fontsize=14, fontweight='bold')
    axes[1].grid(True, alpha=0.3)
    axes[1].set_ylim([0, 1])
    
    plt.tight_layout()
    plt.show()

simulatu_entrenamendua()
print("âœ… Entrenamenduko simulazioa bistaratuta!")
````
